```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
x_values  = df['Year']
y_values  = df['CSIRO Adjusted Sea Level']
```

```python
df
```

# Create scatter plot
 

```python
plt.colormaps()
```

```python
plt.set_cmap("PuRd")
```

```python
fig = plt.scatter(x = x_values,y =y_values , cmap = 'PuRd_r', alpha=0.5)
plt.grid()
plt.figure(figsize =(12,4),  dpi = 100)
plt.show()
```
