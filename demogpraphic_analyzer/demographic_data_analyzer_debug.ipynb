## What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

```python
filter1 = (df['education'] == 'Bachelors' )&(df['salary'] == '>50K')
filter2 = (df['education'] == 'Masters' )&(df['salary'] == '>50K')
filter3 = (df['education'] == 'Doctorate' )&(df['salary'] == '>50K')
```

```python
higher_education=df[(df['education']== 'Bachelors')|(df['education']== 'Masters')|(df['education']== 'Doctorate')]
higher_education
```

```python
my_results = df[(filter1)|(filter2)|(filter3)]
my_result = len(my_results) / len(higher_education)
round(my_result*100,1)
```
